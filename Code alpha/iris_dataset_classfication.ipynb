```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df["target"]=iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df["flower_names"]=df['target'].apply(lambda x:iris.target_names[x])
df.head()
```

```python
df[45:55]
```

```python
df1=df[:50]
df2=df[50:100]
df3=df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python


plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color="black", marker='+')
plt.scatter(df2["sepal length (cm)"], df2["sepal width (cm)"], color="orange", marker="+")

plt.show()

```

```python
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], color="blue", marker='o', label="Class 1")
plt.scatter(df2["petal length (cm)"], df2["petal width (cm)"], color="red", marker="o", label="Class 2")
plt.legend()
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(["target","flower_names"],axis="columns")
y=df.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model=SVC(C=10)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model_g=SVC(gamma=10)
model_g.fit(X_train,y_train)
model_g.score(X_test,y_test)
```

```python
model_linear_kernal=SVC(kernel="linear")
model_linear_kernal.fit(X_train,y_train)
```

```python
model_linear_kernal.score(X_test,y_test)
```
